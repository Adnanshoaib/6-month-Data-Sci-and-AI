# üìÇ Load Dataset

```python
df = pd.read_csv(r"C:\Users\adnan\Downloads\Churn_Modelling.csv")
```

```python
df
```

# üéØ Feature and Target Selection

```python
x = df.iloc[:, 3:-1].values
y = df.iloc[:, -1].values
```

```python
df.head()
```

# üî° Encode Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, 2] = le.fit_transform(x[:, 2])
```

```python
le
```

```python
x
```

# üè∑Ô∏è One-Hot Encode Categorical Features

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder 
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
x
```

# ‚öñÔ∏è Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
sc
```

```python
x
```

# ‚úÇÔ∏è Split Data into Train and Test Sets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)   
print(y_train.shape)
print(y_test.shape)
```

ANN Model

# üß† Build the ANN Model

```python
ann = tf.keras.models.Sequential() # establish LAyer one after another
```

```python
ann
```

```python
ann.add(tf.keras.layers.Dense(units=6)) # input Layer
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu')) # hidden Layer 1
```

```python
#ann.add(tf.keras.layers.Dense(units=5, activation='relu')) # hidden Layer 2
```

```python
#ann.add(tf.keras.layers.Dense(units=4, activation='relu')) # hidden Layer 3
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid')) # output Layer
```

```python
ann
```

# ‚öôÔ∏è Compile the ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

# üöÄ Train the ANN

```python
ann.fit(x_train, y_train, batch_size=32, epochs=150)
```

# üìä Make Predictions

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```
