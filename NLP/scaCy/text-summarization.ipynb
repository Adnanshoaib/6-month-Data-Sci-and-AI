- Imports spaCy, English stopwords, and punctuation symbols.

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

- Creates a list of English stopwords.

```python
stopwords = list(STOP_WORDS)
stopwords
```

```python
len(stopwords)
```

- Loads spaCy’s English NLP model.

```python
nlp = spacy.load('en_core_web_sm')
```

```python
text
```

- Processes the text with spaCy’s NLP model.

```python
doc = nlp(text)
doc
```

- Creates and prints a list of all words (tokens) in the text.

```python
# lets get the tokens from text
tokens = [token.text for token in doc]
print(tokens)
```

```python
tokens
```

```python
len(tokens)
```

```python
punctuation # also called as noisy characters
```

- Calculates the frequency of each non-stopword, non-punctuation word in the text.

```python
#we have to calcualte the freaquency of each and every word, how many time word is repetation in text

word_frequencies = {}

for word in doc:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text] = 1
            else:
                word_frequencies[word.text] += 1
```

```python
word_frequencies
```

```python
len(word_frequencies)
```

```python
word_frequencies
```

 - Finds the highest word frequency in the text.

```python
max_frequency = max(word_frequencies.values())
max_frequency
```

 - Normalizes word frequencies by dividing each by the maximum frequency.

```python
for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word]/max_frequency
```

```python
word_frequencies
```

- Splits the text into sentences and stores them as a list.

```python
sentence_tokens = [sent for sent in doc.sents]
sentence_tokens
```

```python
len(sentence_tokens)
```

```python
sentence_tokens
```

- Calculates a score for each sentence based on the sum of its words’ normalized frequencies.

```python
# we are going to calculate the sentence score, to calculate the sentence score
sentence_scores = {}

for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in word_frequencies.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent] = word_frequencies[word.text.lower()]
            else:
                sentence_scores[sent] += word_frequencies[word.text.lower()]
```

```python
sentence_tokens
```

- Imports a function to get the top N largest elements from a collection.

```python
from heapq import nlargest
```

- Calculates the number of sentences to select (40% of total sentences).

```python
select_length = int(len(sentence_tokens)*0.4)
select_length
```

- Selects the top sentences with the highest scores for the summary.

```python
summary = nlargest(select_length,sentence_scores, key = sentence_scores.get)
```

```python
summary
```

```python
sentence_scores
```

- Creates a list of text from the selected summary sentences.

```python
final_summary = [word.text for word in summary]
```
